```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

# m6A_cdna_d1_vs_d5

```python
d1=pd.read_csv('n2d1_m6A_context.fa-CountByPos-cdna.py.csv', sep='|', index_col=[0])
d5=pd.read_csv('d5-m6A_contexts_to_fasta.py.fa-CountByPos-cdna.py.csv', sep='|', index_col=[0])
d1=d1.drop(['is_greater', 'p'], axis=1)
d1.columns = ['gene', 'length', 'm6A_d1', 'm6A/bp_d1', 'description']
d1['m6A_d5'] = d5['m6A']
d1['m6A/bp_d5'] = d5['m6A/bp']
d1=d1[['gene', 'length', 'm6A_d1', 'm6A/bp_d1','m6A_d5', 'm6A/bp_d5', 'description']]
```

```python
d1.sort_values('m6A/bp_d5', ascending=False)
```

```python
d1.sort_values('m6A/bp_d5', ascending=False).to_excel('m6A_cdna_d1_vs_d5.xlsx')
```
